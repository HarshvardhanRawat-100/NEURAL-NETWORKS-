```python
x = torch.tensor([[1,2,9],[4,5,6]])
print(x)
print(x.shape)
```

```python
# if we want to make a new tensor of same size of previously defined tensor
y = torch.empty_like(x)
print(y)
v = torch.rand_like(x,dtype=float)
print(v)
```

```python
#just checking type
print(x.dtype)
print(y.dtype)
print(v.dtype)
```

```python
#asssign dtype while creating tensor
torch.tensor([0.2,4.3,5.6,0.7],dtype=int)
```

```python
#changing dtype of existing tensor
v = v.to(torch.int)
print(v)
print(v.dtype)
```

```python
# mathematica operations
print(v+2)
print(v*2)
print(v//2)
print(v-2)
```

```python
c = torch.tensor([1,-2,3,-4])
#absolute value of these making these positive
print(torch.abs(c))
#making negative
print(torch.neg(c))
```

```python
d = torch.tensor([0.2,1.9,2.1,4.5,])
#clamp
print(torch.clamp(d,min=2,max=3))
```
