```python
#AUTOGRAD IS A CORE COMPONENT OF PYTORCH THAT PROVIDES AUTOMATIC DIFFERENTIATION FOR TENSOR OPERATIONS
```

```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True)
print(x)
#we use reuires grad to calculate the gradient of that vector afterwards and to track the everything
#requires_grad tell to track the operations for backrop
```

```python
y=x**2
```

```python
y
#here because of require grad we can see y remembers (grad_fn=<PowBackward0>) this thing that there is power backwards
```

```python
#to calculate previous grads
y.backward()
```

```python
#to check those grads
x.grad
```

```python
x = torch.tensor(3.0, requires_grad=True)
y = x**2
z = torch.sin(y)
print(x)
print(y)
print(z)
```

```python
z.backward()
x.grad
```
